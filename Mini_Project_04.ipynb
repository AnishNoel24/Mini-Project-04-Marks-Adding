## **Mini-Project - Marks Adding**

```python
import csv

def modify(csv_file):
    roll_marks = {}
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            roll = row[0]
            marks = int(row[1])
            if roll in roll_marks:
                roll_marks[roll] += 1
            else:
                roll_marks[roll] = 1

    combined_data = [(roll, marks * frequency) for roll, frequency in roll_marks.items()]
    return combined_data


path = '/content/drive/MyDrive/DATASET/Marks Adding.csv'
result = modify(path)

for roll, marks in result:
    print(f"Roll Number: {roll}, Marks: {marks}")
```
